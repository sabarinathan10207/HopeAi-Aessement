```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Salary_Data.csv") #read csv file
```

```python
dataset # view csv
```

```python
independent = dataset[["YearsExperience"]] #Indepedent Variable,years of Experience, datatype float
```

```python
independent
```

```python
dependent = dataset[["Salary"]] # Depedent Variable, Salary, datatype float
```

```python
dependent
```

```python
plt.scatter(independent,dependent)
plt.xlabel('YearsExperience',fontsize=15)
plt.ylabel('Salary',fontsize=15)
plt.show()
```

```python
from sklearn.model_selection import train_test_split #train dataset test split 
X_train, X_test, y_train, y_test = train_test_split(independent,dependent, test_size = 0.05, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)   #y=wx+b
```

```python
weight=regressor.coef_
weight
```

```python
bais=regressor.intercept_
bais
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
filename = 'final model.sav'
pickle.dump(regressor, open(filename, 'wb'))
```
